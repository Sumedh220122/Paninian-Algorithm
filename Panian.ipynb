```python
ach = ['a','i','u','q','L','e','o','E','O','A','I','U','Q']
ik = ['i','u','Q','L']
yan = ['y','v','r','l']
ik_yan = {'i':'y', 'u':'v', 'Q':'r', 'L':'l'}
aech = ['e','o','E','O']
aech_ayav = {'e':'ay','o':'av','E':'Ay','O':'Av'}
khar = ['K', 'P', 'C', 'T', 'W', 'c', 't', 'w', 'k', 'p', 'S', 'R', 's']
```

```python
class Sutra1:

  def ikoyanaachi(self, word):
      new_word = ""
      i = 0
      while i < len(word):
          if word[i] in ik:
              if i != len(word) - 1 and word[i+1] in ach:
                  new_word += ik_yan[word[i]]
                  i += 1
              elif i < len(word) - 2 and word[i+1] == " " and word[i+2] in ach:
                  new_word += ik_yan[word[i]]
                  i += 2
              else:
                  new_word += word[i]
                  i += 1
          else:
              new_word += word[i]
              i += 1
      return new_word

  def aechoayavav(self, word):
    new_word = ""
    i = 0
    while(i < len(word) - 1):
      if word[i] in aech:                            #aechoayavavaha
        if word[i+1] in ach and i != len(word) - 1:
          new_word += aech_ayav[word[i]]
          i = i + 1

        elif word[i+1] == " " and word[i+2] in ach:
          new_word += aech_ayav[word[i]]
          i = i + 3

        else:
          new_word += word[i]
          i = i + 1

      else:
        new_word += word[i]
        i = i + 1

    new_word += word[len(word) - 1]

    return new_word

  def aadgunaha(self, word):
    i = 0
    new_word = ""

    while i < len(word):
        if word[i] == 'a':
            if i != len(word) - 1:
                if word[i+1] in ['i', 'I', 'u', 'U', 'q', 'Q', 'L']:
                    if (i != len(word) - 2 and word[i+1] in ['i', 'I', 'u', 'U', 'q', 'Q']) or (i != len(word) - 3 and word[i+1] == ' '):
                        if word[i+1] in ['i', 'I']:
                            new_word += 'e'
                        elif word[i+1] in ['u', 'U']:
                            new_word += 'o'
                        elif word[i+1] in ['q', 'Q']:
                            new_word += 'ar'
                        elif word[i+1] == 'L':
                            new_word += 'al'
                        i += 2 if word[i+1] != ' ' else 3
                        continue

            new_word += word[i]
            i += 1
        else:
            new_word += word[i]
            i += 1

    return new_word


  def vruddhire(self, word):
      i = 0
      new_word = ""
      while i < len(word) - 1:
        if word[i] == 'a':
          if word[i+1] == 'e' and i != len(word) - 2:
            new_word += 'E'
            i = i + 2

          elif word[i+1] == " " and word[i+2] == 'e':
            new_word += 'E'
            i = i + 3

          elif word[i+1] == 'E' and i != len(word) - 2:
            new_word += 'e'
            i = i + 2

          elif word[i+1] == " " and word[i+2] == 'E':
            new_word += 'e'
            i = i + 3

          elif word[i+1] == 'o' and i != len(word) - 2:
            new_word += 'O'
            i = i + 2

          elif word[i+1] == " " and word[i+2] == 'o':
            new_word += 'O'
            i = i + 3

          elif word[i+1] == 'O' and i != len(word) - 2:
            new_word += 'O'
            i = i + 2

          elif word[i+1] == " " and word[i+2] == 'O':
            new_word += 'O'
            i = i + 3

          else:
            new_word += word[i]
            i = i + 1

        else:
            new_word += word[i]
            i = i + 1

      new_word += word[len(word) - 1]
      return new_word

  def engapada(self, word):
        result = []
        i = 0

        while i < len(word) - 1:
            if (word[i] == 'e' and word[i+1] == 'a') or (word[i] == 'o' and word[i+1] == 'a'):
                result.append(word[i])
                i += 2

            elif (word[i] == 'e' and word[i+1] == ' ' and word[i+2] == 'a') or (word[i] == 'o' and word[i+1] == ' ' and word[i+2] == 'a'):
                result.append(word[i])
                i = i + 3
            else:
                result.append(word[i])
                i += 1

        if i == len(word) - 1:
            result.append(word[i])

        new_word = ''.join(result)

        return new_word

  def savarnaderg(self, word):
    new_word = ""
    i = 0

    word_dict = {'a': 'A', 'i': 'I', 'u' : 'U', 'q': 'Q'}

    while(i < len(word)):
      if i < len(word) - 2:
        if word[i] == 'a' and word[i+1] == 'a':
          new_word += word_dict['a']
          i = i + 2
        elif word[i] == 'a' and word[i+1] == " " and word[i+2] == 'a':
          new_word += word_dict['a']
          i = i + 3

        elif word[i] == 'i' and word[i+1] == 'i':
          new_word += word_dict['i']
          i = i + 2
        elif word[i] == 'i' and word[i+1] == " " and word[i+2] == 'i':
          new_word += word_dict['i']
          i = i + 3

        elif word[i] == 'u' and word[i+1] == 'u':
          new_word += word_dict['U']
          i = i + 2
        elif word[i] == 'u' and word[i+1] == " " and word[i+2] == 'u':
          new_word += word_dict['u']
          i = i + 3

        elif word[i] == 'q' and word[i+1] == 'q':
          new_word += word_dict['q']
          i = i + 2
        elif word[i] == 'q' and word[i+1] == " " and word[i+2] == 'q':
          new_word += word_dict['q']
          i = i + 3

        else:
          new_word += word[i]
          i = i + 1
      else:
        new_word += word[i]
        i = i + 1

    return new_word
```

```python
class Sutra2:
  def stohoschunaschuhu(self, word):
    new_word = ""
    my_dict = {'s': 'S', 'w': 'c', 'W': 'C', 'x': 'j', 'X': 'J', 'n': 'F'}
    target = ['S', 'C', 'c', 'j', 'J', 'F']
    i = 0
    while i < len(word):
        if i < len(word) - 1:
            if word[i] in my_dict and word[i + 1] in target:
                new_word += my_dict.get(word[i])
                i += 1
            elif word[i] in my_dict and word[i - 1] in target:
                new_word += my_dict.get(word[i])
                i += 1
            else:
                new_word += word[i]
                i += 1
        else:
            new_word += word[i]
            i += 1

    return new_word


  def shtunashtuhu(self, word):
    i = 0
    new_word = ""
    my_dict = {'s': 'R', 'w': 't', 'W': 'T', 'x': 'd', 'X': 'D', 'n': 'N'}
    target = ['R', 't', 'T', 'd', 'D', 'N']

    while i < len(word):
        if i < len(word) - 2:
            if (i == 0 and word[i] in my_dict and word[i + 1] in target):
                new_word += my_dict.get(word[i])
                i += 1
            elif word[i] in my_dict and word[i - 1] in target:
                new_word += my_dict.get(word[i])
                i += 1
            elif (word[i] in my_dict and word[i + 1] in target):
                new_word += my_dict.get(word[i])
                i += 1
            else:
                new_word += word[i]
                i += 1
        else:
            new_word += word[i]
            i += 1

    return new_word



  def sasaishoruhu(self, word):
      new_word = ""
      i = 0

      while(i < len(word)-1):
        if word[i] == 's' and word[i+1] == " " and i != 0:
          new_word += 'r'
          i += 2
        else:
          new_word += word[i]
          i += 1

      new_word += word[len(word) - 1]


      return new_word

  def hashicha(self, word):
      i = 0
      new_word = ""

      hash = ['h', 'y', 'v', 'r', 'l', 'F', 'f', 'N', 'n', 'J', 'B', 'g', 'd', 'x']

      while(i < len(word) - 1):
        if word[i] == 'r' and i != 0:
          if word[i-1] == 'a' and word[i+1] in hash:
            new_word += 'u'
            i = i + 1
          else:
            new_word += word[i]
            i = i + 1
        else:
          new_word += word[i]
          i = i + 1

      new_word += word[len(word) - 1]

      return new_word


  def rori(self, word):
      i = 0
      new_word = ""

      while(i < len(word) - 1):
        if word[i] == 'r' and word[i+1] == 'r':
           i = i + 1
        else:
          new_word += word[i]
          i = i + 1

      new_word += word[len(word) - 1]

      return new_word
```

```python
class Sutra3:
  def kharavasana(self, word):
    end = len(word) - 1
    i = 0
    new_word = ""


    if word[end] == 'r':
      new_word += word[0: end] + 'aH'

    if word[end] == 's':
      new_word += word[0: end] + 'H'

    else:
      while(i < len(word)):
       if i < len(word) - 2:
          if word[i] == 'r' and word[i+1] in khar:
            new_word += 'aH'
            i= i + 1
          elif word[i] == 'r' and word[i+1] == " " and word[i+2] in khar:
            new_word += 'aH'
            i= i + 1
          else:
            new_word += word[i]
            i = i + 1
       else:
          new_word += word[i]
          i = i + 1


    return new_word

  def visarjaniyasya(self, word):
    i = 0
    new_word = ""

    while(i < len(word) - 2):
      if word[i] == 'a' and word[i+1] == 'H' and word[i+2] in khar:
        new_word += 's'
        i = i + 2
      else:
        new_word += word[i]
        i = i + 1

    new_word += word[-2] + word[-1]

    return new_word


  def vashari(self, word):
    i = 0
    new_word = ""

    while(i < len(word) - 2):
      if word[i] == 'a' and word[i+1] == 'H' and word[i+2] in ['S', 'R', 's']:
        new_word += 's'
        i = i + 2
      else:
        new_word += word[i]
        i = i + 1

    new_word += word[-2] + word[-1]

    return new_word


  def custom(self, word):
    i = 0
    new_word = ""

    while(i < len(word)):
      if i < len(word) - 2:
        if (word[i] == 'a' or word[i] == 'A') and word[i+1] == 'A':
          i += 1
        elif (word[i] == 'a' or word[i] == 'A') and word[i+1] == " " and word[i+2] == 'A' and i + 2 < len(word):
          i += 2

        elif (word[i] == 'i' or word[i] == 'I') and word[i+1] == 'I':
          i += 1
        elif (word[i] == 'i' or word[i] == 'I') and word[i+2] == 'I' and i + 2 < len(word):
          i += 2

        elif (word[i] == 'u' or word[i] == 'U') and word[i+1] == 'U':
          i += 1
        elif (word[i] == 'u' or word[i] == 'U') and word[i+1] == " " and word[i+2] == 'U' and i + 2 < len(word):
          i += 2

        elif (word[i] == 'q' or word[i] == 'Q') and word[i+1] == 'Q':
          i += 1
        elif (word[i] == 'q' or word[i] == 'Q') and word[i+1] == " " and word[i+2] == 'Q' and i + 2 < len(word):
          i += 2

        else:
          new_word += word[i]
          i += 1
      else:
        new_word += word[i]
        i += 1
    return new_word
```
